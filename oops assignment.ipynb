Q3. Explain why the __init__() function is used. Give a suitable example.
Ans. The __init__() function is a special method in Python classes that is automatically called when an object is created from the class. It is commonly used to initialize the attributes of the object with values passed as arguments during object creation.

The primary purpose of the __init__() function is to ensure that the newly created object is properly initialized and in a valid state before it is used.

Here's an example to illustrate the usage of the __init__() function:
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

person1 = Person("Alice", 30)

we have a Person class with two attributes: name and age. The __init__() method is defined with self as the first parameter, which refers to the object being created. It also takes two additional parameters: name and age.

When we create an object of the Person class using the line person1 = Person("Alice", 30), the __init__() method is automatically called with the values "Alice" and 30 for the name and age parameters, respectively. Inside the __init__() method, these values are assigned to the corresponding attributes of the object using self.name = name and self.age = age.

As a result, the person1 object is initialized with the provided values for name and age. These attributes can then be accessed and used by other methods or code that interacts with the person1 object.

The __init__() function allows us to provide initial values for the object's attributes during object creation, ensuring that the object starts in a valid state with the desired initial values.
 

Q4. Why self is used in OOPs?  

Ans. In object-oriented programming (OOP), the self keyword is used as a convention to refer to the instance of a class within its methods. It acts as a reference to the current object on which the method is being called.

The use of self is important because it allows you to access and modify the attributes and methods of the current object. By convention, the first parameter of a class method in Python is always self, although you can give it any name you like (but using self is strongly recommended to follow the convention and make your code more readable for others).

Here are a few reasons why self is used in OOP:

Object-specific access: self provides a way to access and manipulate the attributes and methods of the current object. Each object created from a class has its own set of attributes, and self allows you to refer to those specific attributes within the methods of the class.

Differentiating instance variables from local variables: In a class method, there can be both instance variables (defined in the class) and local variables (defined within the method). By using self, you can clearly differentiate between the two and avoid any confusion or naming conflicts.

Method invocation: When a method is called on an object, self allows the method to refer to the specific instance of the object on which the method is invoked. It ensures that the method operates on the correct object and can access its specific data.

Object-oriented nature: self promotes the principles of encapsulation and abstraction in OOP. It emphasizes that methods and attributes are tied to individual objects rather than being global or shared across all instances of a class.


class Person:
    def __init__(self, name):
        self.name = name
    
    def greet(self):
        print(f"Hello, my name is {self.name}.")

person1 = Person("Alice")
person1.greet()  # Output: Hello, my name is Alice.


the self.name attribute is specific to each instance of the Person class. The self keyword allows the greet() method to access the name attribute of the specific person1 object. Without self, the method would not know which object's name attribute to refer to.

Q5. What is inheritance? Give an example for each type of inheritance.

Ans.Inheritance is a fundamental concept in object-oriented programming (OOP) that allows a class to inherit attributes and methods from another class. The class that inherits is called the child class or derived class, and the class being inherited from is called the parent class or base class.

Single Inheritance:
class Animal:
    def __init__(self, name):
        self.name = name
    
    def eat(self):
        print("The animal is eating.")
    
    def sleep(self):
        print("The animal is sleeping.")

class Dog(Animal):
    def bark(self):
        print("The dog is barking.")

dog = Dog("Buddy")
print(dog.name)  # Output: Buddy
dog.eat()        # Output: The animal is eating.
dog.sleep()      # Output: The animal is sleeping.
dog.bark()       # Output: The dog is barking.

Multiple Inheritance:
class FlyingBird:
    def fly(self):
        print("The bird is flying.")
    
    def buildNest(self):
        print("The bird is building a nest.")

class SwimmingFish:
    def swim(self):
        print("The fish is swimming.")
    
    def layEggs(self):
        print("The fish is laying eggs.")

class Duck(FlyingBird, SwimmingFish):
    def quack(self):
        print("The duck is quacking.")

duck = Duck()
duck.fly()         # Output: The bird is flying.
duck.buildNest()   # Output: The bird is building a nest.
duck.swim()        # Output: The fish is swimming.
duck.layEggs()     # Output: The fish is laying eggs.
duck.quack()       # Output: The duck is quacking.


Multilevel Inheritance:
class Vehicle:
    def start(self):
        print("The vehicle is starting.")
    
    def stop(self):
        print("The vehicle is stopping.")

class Car(Vehicle):
    def drive(self):
        print("The car is driving.")

class SportsCar(Car):
    def accelerate(self):
        print("The sports car is accelerating.")

sportsCar = SportsCar()
sportsCar.start()       # Output: The vehicle is starting.
sportsCar.drive()       # Output: The car is driving.
sportsCar.accelerate()  # Output: The sports car is accelerating.
sportsCar.stop()        # Output: The vehicle is stopping.


Hierarchical Inheritance:

class Shape:
    def calculateArea(self):
        print("Calculating area.")

    def draw(self):
        print("Drawing shape.")

class Circle(Shape):
    def calculateArea(self):
        print("Calculating area of circle.")

class Rectangle(Shape):
    def calculateArea(self):
        print("Calculating area of rectangle.")

class Triangle(Shape):
    def calculateArea(self):
        print("Calculating area of triangle.")

circle = Circle()
circle.calculateArea()  # Output: Calculating area of circle.
circle.draw()           # Output: Drawing shape.

rectangle = Rectangle()
rectangle.calculateArea()  # Output: Calculating area of rectangle.
rectangle.draw()           # Output: Drawing shape.

triangle = Triangle()
triangle.calculateArea()   # Output: Calculating area of triangle.
triangle.draw()            # Output: Drawing shape.
